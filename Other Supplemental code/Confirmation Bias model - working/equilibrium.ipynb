```python
import numpy as np
import xlrd
import matplotlib.pyplot as plt

##generate adjacency matrix
file_location = "C:/Users/Mao/Desktop/Competitive Social Paper/Python Codes/friendship.xlsx"
workbook = xlrd.open_workbook(file_location)
sheet = workbook.sheet_by_index(0)
nrows = sheet.nrows  
ncols = sheet.ncols 
tere = np.arange(0, nrows, 1)
adjacency = np.zeros((nrows, ncols))

for x in range(ncols):
    cols = sheet.col_values(x)
    adjacency[:, x] = np.matrix(cols)
adjacency = np.transpose(adjacency)    
for i in range(nrows):
    degree = sum(adjacency[i,:])
    for j in range(nrows):
        adjacency[i,j] = (adjacency[i,j])/(degree+15)
##

##calculate eigenvector centralities and maximum eigenvalue
adjacency_transpose = np.transpose(adjacency)
egva,egve = np.linalg.eig(adjacency_transpose)
eigvector_centralities = egve[:,0]
eigenvalue_max = max(egva)
##

##generate group innate opinions that satisfies assuption
s_min = 0.3  
s_max = 0.7   
temp_innate = np.arange(s_min,s_max,(s_max-s_min)/20)
innate = np.append(temp_innate,[s_max],axis=0)
rowsum = np.sum(adjacency,1)
pre_mean = (sum(eigvector_centralities*innate))/sum(eigvector_centralities)
differnece = ((0.5 - pre_mean)*sum(eigvector_centralities))/eigvector_centralities[18]
innate[18] += np.real(differnece)
innate_mean = (sum(eigvector_centralities*innate))/sum(eigvector_centralities)
##

##generate parameteres of conformation bias
beta = 0.2;
gamma = 0.1
##

##system dynamics
g = 0.2
h = 0.8
x = np.random.rand(nrows,1)
opinion = x
running_time = 2000
for t in range(running_time):
    alpha = np.zeros((nrows, ncols))
    tem_alpha = np.ones((nrows,1)) - np.transpose([rowsum]) - (beta*(np.ones((nrows,1)))- gamma*(np.absolute(x-g*np.ones((nrows,1))))) \
                                                            - (beta*(np.ones((nrows,1)))- gamma*(np.absolute(x-h*np.ones((nrows,1)))))
    for i in range(nrows):
        alpha[i,i] = tem_alpha[i]  
    x = np.matmul(alpha,np.transpose([innate])) + np.matmul(adjacency,x) \
                                                + (beta*(np.ones((nrows,1)))- gamma*(np.absolute(x-g*np.ones((nrows,1)))))*g \
                                                + (beta*(np.ones((nrows,1)))- gamma*(np.absolute(x-h*np.ones((nrows,1)))))*h
    opinion = np.append(opinion, x,1)
##

##calculate equlibriua manually (Successful!!!)
rowsumm_atrix = np.zeros((nrows, ncols))
for i in range(nrows):
    rowsumm_atrix[i,i] = rowsum[i]
Ematrix = np.eye(nrows) - adjacency + (g-h)*gamma*(np.eye(nrows))
Ematrix_inverse = np.linalg.inv(Ematrix)
addMatrix = np.eye(nrows) - rowsumm_atrix - 2*beta*(np.eye(nrows)) + (h-g)*gamma*(np.eye(nrows))
Matrix_A = np.matmul(Ematrix_inverse,addMatrix)
Vector_A = (h+g)*beta*(np.ones((nrows,1))) + ((g**2)-(h**2))*gamma*(np.ones((nrows,1)))
equilibriua = np.matmul(Matrix_A,np.transpose([innate])) + np.matmul(Ematrix_inverse,Vector_A)
##

##unity function
unity = np.matmul(eigvector_centralities,equilibriua)
print(unity)
##





#print(x)
#maxeg = max(egva)
#inopinion = np.transpose([innate])
#print(maxeg)
#print(ass1)
##print(egve[:,0])
##verification for convergence



#rt = np.arange(21)
#plt.plot(rt,equilibrium,'r')
#plt.show()








```
